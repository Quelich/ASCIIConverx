```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Defining a neural network with one layer and one neuron
model = Sequential([Dense(units=1, input_shape=[1])])

# Measuring how good or how bad that guess is
# So that making another "optimized" guess
model.compile(optimizer='sgd', loss='mean_squared_error')

# Formatting the the numbers into the data format that the layers expect
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0, ], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0, ], dtype=float)

# Learning process begins
model.fit(xs, ys, epochs=500)

result = model.predict([10.0])

print(result)
```
